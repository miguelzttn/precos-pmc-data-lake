```python

gold_tables = [
    'f_precos',
    'f_precos_completa',
    'd_empresas',
    'd_produtos'
]

base = "s3a://precos-pmc-app-db/gold_parquet"

for table in gold_tables:
    name = table.split(".")[-1]
    print(f'Processing: {name}')
    (
        spark.table(f'{SCHEMA}.{CATALOG}.{table}')
        .write
        .mode("overwrite")
        .parquet(f"{base}/{name}")
    )
```
