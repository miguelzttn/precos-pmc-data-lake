```python
from pyspark.sql.functions import regexp_extract, col, last, first
```


## Produto - Dicionário de dados


```python

path_dicionarios_produtos = (
    "s3://precos-pmc-landing"
    "/*" # YYYY
    "/*" # MM
    "/*" # DD
    "/*_Clique_Economia_-_Produto_-_Dicionario_de_Dados.csv"
)

df_dicionarios_produtos = (
    spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .csv(path_dicionarios_produtos)
        
        # Extract filename
        .withColumn("file_path", col("_metadata.file_path"))
        .withColumn(
            "filename",
            regexp_extract(col("_metadata.file_path"), r"([^/]+$)", 1)
        )
        .drop("file_path")
)

df_dicionarios_produtos = (
    df_dicionarios_produtos
        #.OrderBy("filename")
        .groupBy([name for name in df_dicionarios_produtos.columns if name != 'filename'])
        .agg(
            last("filename").alias("last_filename"),
            first("filename").alias("first_filename"), 
        )
)

display(df_dicionarios_produtos)
```


Não há nenhuma informação que leve a crer que houve alteração no schema da base de dados de produtos. Tratarei como uma só.
